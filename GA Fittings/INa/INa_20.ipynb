```python
import timeit
import numpy as np
import matplotlib.pyplot as plt
import INa_worker
from multiprocessing import Pool
# from concurrent import futures
from sklearn.metrics import r2_score
import pickle
import warnings
warnings.filterwarnings('ignore')
# np.random.seed(10000)

Runtime_arr = []
class GA_trial:
    gen_num = 0
    loss = 0
    R_sq_ini = 0
    R_sq_fin = 0
    Parameters = 0
    final_fit_Val = 0
    time = 0
    cur_time_fit = 0
    cur_time_ini = 0
    Vcm = 0
    I_peaks_fin = 0
    I_peaks_ini = 0
    Runtime = 0

ini_par = np.array([18, 0.32, 0.08, 0.135, 3.56, 3.1e5, 1.2714e5, 0.2444,
                    3.474e-5, 0.04391, 0.311, 0.1212, 0.01052, 0.13, 10.66, 11.1, 0.3, 2.535e-7, 0.1, 32])
INa_IVini = INa_worker.INa_func(ini_par)
time_ini, plt_INa_ini, size = INa_worker.INa_func_time(ini_par)

timer_start = timeit.default_timer()

V = np.arange(-80, 40, 5)

INa_act = 7*np.array(
    [-0.181892104, -0.254088538, -0.244609315, -0.555747839, -1.305390034, -3.748475601, -11.94922772, -29.08036381
        , -47.19266789, -60.86300032, -69.75161538, -72.47572059, -71.52890259, -67.76456532, -61.12362642, -54.03297911
        , -45.8534729, -37.46637645, -27.97033381, -18.02570752, -5.800137618, 2.690300795, 8.969040799, 19.13778239
     ])

pop_size = 27*3

# # fitness
def pop_fitness(ppln):
    p = Pool(processes = 28)
    INa_cur = p.map(INa_worker.INa_func, ppln)
    SSE = np.sqrt(np.sum(np.power((INa_cur - INa_act), 2, dtype=np.float64), axis=1)/len(V)) #RMSE
    return SSE

def select_mating_pool(popn, fitness, num_parents):
    parents = np.empty((num_parents, popn.shape[1]))
    for parent_no in range(num_parents):
        max_fit_idx = np.where(fitness == np.min(fitness))  # find max and return its index
        max_fit_idx = max_fit_idx[0][0]  # Select the first index if repetitive
        parents[parent_no, :] = popn[max_fit_idx, :]
        fitness[max_fit_idx] = +9999999999999
    return parents

# 4-point crossover
def crossover(parents,
              offspring_size):  # in the case offspring size = (k, 3) where k is the number of desired offsprings
    offspring = np.empty(offspring_size)
    crossover_point1 = np.uint8(np.random.randint(0, offspring_size[1]/4))
    crossover_point2 = np.uint8(np.random.randint(offspring_size[1]/4, offspring_size[1]*2/4))
    crossover_point3 = np.uint8(np.random.randint(offspring_size[1]*2/4, offspring_size[1]*3/4))
    crossover_point4 = np.uint8(np.random.randint(offspring_size[1]*3/4, offspring_size[1]))

    for n in range(parents.shape[0]):
        # parent selection
        parent1_idx = n % parents.shape[0]  # using modulus helps implement a ring parent selection fashion
        parent2_idx = (n + 1) % parents.shape[0]
        
        #first offspring
        offspring[n, 0:crossover_point1] = parents[parent1_idx, 0:crossover_point1]
        offspring[n, crossover_point1:crossover_point2] = parents[parent2_idx, crossover_point1:crossover_point2]
        offspring[n, crossover_point2:] = parents[parent1_idx, crossover_point2:]
        #second offspring
        offspring[n + parents.shape[0], 0:crossover_point3] = parents[parent1_idx, 0:crossover_point3]
        offspring[n + parents.shape[0], crossover_point3:] = parents[parent2_idx, crossover_point3:]
        #third offspring
        offspring[n + 2*parents.shape[0], 0:crossover_point1] = parents[parent2_idx, 0:crossover_point1]
        offspring[n + 2*parents.shape[0], crossover_point1:crossover_point2] = parents[parent1_idx, crossover_point1:crossover_point2]
        offspring[n + 2*parents.shape[0], crossover_point2:crossover_point3] = parents[parent2_idx, crossover_point2:crossover_point3]
        offspring[n + 2*parents.shape[0], crossover_point3:crossover_point4] = parents[parent1_idx, crossover_point3:crossover_point4]
        offspring[n + 2*parents.shape[0], crossover_point4:] = parents[parent2_idx, crossover_point4:]
        #fourth offspring
        offspring[n + 3*parents.shape[0], 0:crossover_point4] = parents[parent2_idx, 0:crossover_point4]
        offspring[n + 3*parents.shape[0], crossover_point4:] = parents[parent2_idx, crossover_point4:]
    return offspring


# single gene mutation per offspring
def mutation(offspring_crossover, best_par):
    gen_alpha = np.random.uniform(0.0, 0.2, C.shape[1])
    for idx in range(offspring_crossover.shape[0]):
        gene_select = np.random.randint(0, offspring_crossover.shape[1])  # vary range of random value for gene variation
        rnd_single_gene_mutn_val = 0
        for i in range(offspring_crossover.shape[1]):
            rnd_single_gene_mutn_val = rnd_single_gene_mutn_val + (gene_select == i)*np.random.uniform(-gen_alpha[i]*best_par[i], gen_alpha[i]*best_par[i])
        offspring_crossover[idx, gene_select] = offspring_crossover[idx, gene_select] + rnd_single_gene_mutn_val
    return offspring_crossover

# num_of_trials = 1
trial_count = 0

timer_start = timeit.default_timer()
C = np.empty((pop_size, len(ini_par)))
pop_alpha = np.random.uniform(0.0, 0.8, C.shape[1])
for i in range(C.shape[1]):
    C[:, i] = np.random.uniform((1 - pop_alpha[i])*ini_par[i], (1 + pop_alpha[i])*ini_par[i], C.shape[0])

INa = np.empty((C.shape[0], len(V)))
population = C  # initial Population

print('GA Trial %s' %trial_count)
print('*************************************************************************************')

i_plt = 0
gen_mod = 10
num_of_gens = 101
no_plts = num_of_gens/gen_mod
no_plts = np.int16(no_plts)
gen_curr_arr = np.zeros((len(V), num_of_gens))
num_mating_parents = int(C.shape[0]/9)
best_gen_fitness = []

for generation in range(num_of_gens):
    fitness = pop_fitness(population)  # calculate fitness (SSE)
    #print(fitness)
    fitness = np.where(np.isnan(fitness), 1e120, fitness) #handles the dreaded nan's
    fitness = np.where(np.isinf(fitness), 1e120, fitness)
    best_gn_fit_idx = np.where(fitness == np.min(fitness))
    indx = best_gn_fit_idx[0][0]
    best_gen_fitness.append(fitness[indx])

    if generation%gen_mod == 0:
        gen_num = np.arange(len(best_gen_fitness))
        timer_stop = timeit.default_timer()
        print('generation: ', generation)
        Rtime = (timer_stop - timer_start)/60.0
        print('Runtime: ', Rtime , 'mins')
        print('Best fitness: ', fitness[indx])
        Runtime_arr.append(Rtime)

        y_fit = INa_worker.INa_func(population[indx])
        time, plot_INa_fit, size = INa_worker.INa_func_time(population[indx])
        t, s, m, h, j = INa_worker.mhj(population[indx])
        gen_curr_arr[:, i_plt] = y_fit
        print('generational solution: ',population[indx])
        plt.figure(figsize=(16,4))
        plt.subplot(131)
        plt.plot(V, INa_act, 'r-*', V, y_fit, '-o')
        plt.legend(['Experimental', 'Model-fitted'], loc='best')
        plt.xlabel('Command Voltage, Vc (mV)')
        plt.ylabel('Current Density (pA/pF)')
        plt.title('$I_{Na}$ IV plot: Experimental vrs Grandi-Pandit Model fitted')
        plt.grid()

        plt.subplot(132)
        plt.grid()
        plt_cnt = -80
        for k in np.arange(len(V)):     
            plt.plot(time[k, : size[k]], plot_INa_fit[k, : size[k]], label='%s mV' %plt_cnt)
            plt_cnt = plt_cnt+5
        plt.xlabel('time (ms)')
        plt.ylabel('Current Density (pA/pF)')
#         plt.legend(loc='lower right')
        plt.title('Time course of $I_{Na}$')
        
        plt.subplot(133)
        for k in np.arange(len(V)):
            plt.plot(t[k, : s[k]], m[k, : s[k]], '-r')
        for k in np.arange(len(V)):
            plt.plot(t[k, : s[k]], h[k, : s[k]], '-g')
        for k in np.arange(len(V)):
            plt.plot(t[k, : s[k]], j[k, : s[k]], '-b')
        plt.xlabel('Time (ms)')
        plt.title('gating (m-red, h-green, j-blue)')
        plt.tight_layout()
        plt.show()

    if generation == num_of_gens-1:
        Best_Sol = population[indx]
        final_fitness = fitness[indx]
        print('-------------------------------------------------------------------')
        print('Best Solution: ', Best_Sol)
        print('Final Fitness: ', final_fitness)
        ini_r2 = r2_score(INa_act, INa_IVini)
        fit_r2 = r2_score(INa_act, y_fit)
        plt.figure()
        plt.plot(V, INa_act, 'r-*', V, y_fit, '-o')
        plt.legend(['Experimental', 'Model-fitted'], loc='best')
        plt.xlabel('Command Voltage, Vc (mV)')
        plt.ylabel('Current Density (pA/pF)')
        plt.title('$I_{Na}$ IV plot: Experimental vrs Grandi-Pandit Model fitted')
        plt.grid()
        plt.show()

        plt.figure(figsize=(10,8))
        for j in np.arange(len(V)):
            plt.plot(time[j, : size[j]], plot_INa_fit[j, : size[j]])
        plt.xlabel('time (ms)')
        plt.ylabel('Current Density (pA/pF)')
        plt.title('Time course of $I_{Na}$')
#         plt.grid()
        plt.show()

    parents = select_mating_pool(population, fitness, num_mating_parents)  # choose mating parents based on best min SSE
    offspring_crossed = crossover(parents, offspring_size=(4*parents.shape[0], population.shape[1]))
    offspring_mutated = mutation(offspring_crossed, population[indx])
    population[0:parents.shape[0], :] = parents
    population[parents.shape[0]: 5 * parents.shape[0], :] = offspring_crossed
    population[5 * parents.shape[0]:, :] = offspring_mutated

    i_plt = i_plt + 1

timer_stop = timeit.default_timer()        
print('Total Runtime: ', (timer_stop - timer_start)/60.0, 'mins')

gen_num = np.arange(len(best_gen_fitness))

#fitted model
INa_IVfit = INa_worker.INa_func(Best_Sol)
time_fit, plt_INa_fit, size = INa_worker.INa_func_time(Best_Sol)

print('ini_r2: ', ini_r2)
print('fit_r2: ', fit_r2)

GA_trial_n = GA_trial()
GA_trial_n.gen_num = gen_num
GA_trial_n.loss = best_gen_fitness
GA_trial_n.final_fit_val = final_fitness
GA_trial_n.R_sq_ini = ini_r2
GA_trial_n.R_sq_fin = fit_r2
GA_trial_n.Parameters = Best_Sol
GA_trial_n.time = time
GA_trial_n.cur_time_ini = plt_INa_ini
GA_trial_n.cur_time_fit = plt_INa_fit
GA_trial_n.Vcm = V
GA_trial_n.gen_fit_arr = gen_curr_arr
GA_trial_n.I_peaks_fin = y_fit
GA_trial_n.I_peaks_ini = INa_IVini
GA_trial_n.Runtime = Runtime_arr

pickle.dump(GA_trial_n, open('Params/INa_LRd20_GA_trial%s.p' %trial_count, 'wb'))

gen_num = np.arange(len(best_gen_fitness))
plt.figure()
plt.plot(gen_num, best_gen_fitness)
plt.yscale('log')
plt.xlabel('number of generations')
plt.ylabel('Loss: Sum of Squared Error (log scale)')
plt.grid()
plt.show()

plt.figure(figsize=(7,6))
plt.plot(V, INa_act, 'b-o', label = 'expt')
for cnt in np.arange(0,gen_curr_arr.shape[1], gen_mod):
    plt.plot(V, gen_curr_arr[:, cnt], '-*', label='gen %s' %cnt)

plt.xlabel('Transmembrane Voltage, $V_m$ (mV)')
plt.ylabel('Current Density (pA/ pF)')
plt.title('$I_{Na}$ IV plots - Experimental vs multiple generations')
plt.grid()
plt.legend(loc='best')
plt.show()

plt.figure()
for j in np.arange(len(V)):
    plt.plot(time[j, : size[j]], plt_INa_fit[j, : size[j]])
plt.xlabel('time(ms)')
plt.ylabel('Current Density (pA/pF)')
plt.title('Time Course of $I_{Na}$ under various clamp voltages')
plt.grid()
plt.show()
```
